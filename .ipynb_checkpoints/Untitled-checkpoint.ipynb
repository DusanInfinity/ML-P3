Booking_ID: unique identifier of each booking
no_of_adults: Number of adults
no_of_children: Number of Children
no_of_weekend_nights: Number of weekend nights (Saturday or Sunday) the guest stayed or booked to stay at the hotel
no_of_week_nights: Number of week nights (Monday to Friday) the guest stayed or booked to stay at the hotel
type_of_meal_plan: Type of meal plan booked by the customer:
required_car_parking_space: Does the customer require a car parking space? (0 - No, 1- Yes)
room_type_reserved: Type of room reserved by the customer. The values are ciphered (encoded) by INN Hotels.
lead_time: Number of days between the date of booking and the arrival date
arrival_year: Year of arrival date
arrival_month: Month of arrival date
arrival_date: Date of the month
market_segment_type: Market segment designation.
repeated_guest: Is the customer a repeated guest? (0 - No, 1- Yes)
no_of_previous_cancellations: Number of previous bookings that were canceled by the customer prior to the current booking
no_of_previous_bookings_not_canceled: Number of previous bookings not canceled by the customer prior to the current booking
avg_price_per_room: Average price per day of the reservation; prices of the rooms are dynamic. (in euros)
no_of_special_requests: Total number of special requests made by the customer (e.g. high floor, view from the room, etc)
booking_status: Flag indicating if the booking was canceled or not.

```python
import tensorflow as tf
from tensorflow.keras import layers, Sequential
from IPython.display import Image
from tensorflow.keras.utils import plot_model
from keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.utils.class_weight import compute_class_weight
from tensorflow.keras import backend as K

def custom_loss(y_true, y_pred):
    y_true = tf.cast(y_true, y_pred.dtype)
    # Weight for the penalty
    penalty_weight = 0.8      
    # Binary cross-entropy loss
    binary_crossentropy_loss = K.binary_crossentropy(y_true, y_pred)    
    # Penalty for misses on class 0
    class_0_penalty = penalty_weight * K.mean(K.square(y_true - y_pred) * (1 - y_true), axis=-1)    
    # Total loss
    loss = binary_crossentropy_loss + class_0_penalty    
    return loss


def custom_loss2(y_true, y_pred):
    y_true = tf.cast(y_true, y_pred.dtype)    
    # Binary cross-entropy loss
    binary_crossentropy_loss = K.binary_crossentropy(y_true, y_pred)    
    # Weight for the penalties
    class_0_penalty_weight = 0.8
    class_1_penalty_weight = 1.0
    false_positive_penalty_weight = 1.2
    false_negative_penalty_weight = 1.5    
    # Penalty for misses on class 0
    class_0_penalty = class_0_penalty_weight * K.mean(K.square(y_true - y_pred) * (1 - y_true), axis=-1)    
    # Penalty for misses on class 1
    class_1_penalty = class_1_penalty_weight * K.mean(K.square(y_true - y_pred) * y_true, axis=-1)    
    # Additional penalty term for false positives
    false_positive_penalty = false_positive_penalty_weight * K.mean(K.square(tf.maximum(y_pred - y_true, 0)), axis=-1)    
    # Additional penalty term for false negatives
    false_negative_penalty = false_negative_penalty_weight * K.mean(K.square(tf.maximum(y_true - y_pred, 0)), axis=-1)    
    # Total loss
    loss = binary_crossentropy_loss + class_0_penalty + class_1_penalty + false_positive_penalty + false_negative_penalty    
    return loss


# Assuming X_train is your training features and y_train is your training labels
input_shape = X_train.shape[1]  # Number of features in your training data

# Define the model
model = Sequential([
    layers.Dense(16, activation='relu', input_shape=(input_shape,)),
    layers.Dropout(0.2),  # Adding dropout for regularization
    layers.Dense(16, activation='relu'), 
    
    #kernel_regularizer=regularizers.l2(0.01),
    
    
    layers.BatchNormalization(),  # Adding batch normalization
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.3),  # Adding dropout for regularization
    layers.Dense(16, activation='relu'),
    layers.Dense(1, activation='sigmoid')  # Output layer with sigmoid activation for binary classification
])

# Compile the model
model.compile(optimizer='adam',
              #loss=custom_loss,  # For binary classification
              loss = custom_loss2,
              metrics=['accuracy'])


callbacks = [EarlyStopping(monitor="val_loss", patience=2),
 ModelCheckpoint(filepath="best_model.h5",
 monitor="val_loss",
save_best_only=True)]


# Fit the model with early stopping
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2, callbacks=[early_stopping])

# Evaluate the model on your test data
test_loss, test_accuracy = model.evaluate(X_test, y_test)

print("Test Loss:", test_loss)
print("Test Accuracy:", test_accuracy)



model.summary()
#binary_crossentropy mse
```

```python
from sklearn.model_selection import StratifiedKFold

# Define number of folds
n_folds = 3

# Initialize StratifiedKFold
skf = StratifiedKFold(n_splits=n_folds, shuffle=True, random_state=42)

# Initialize lists to store evaluation results
losses = []
accuracies = []

# Loop through each fold
for train_index, val_index in skf.split(X_train, y_train):
    # Create model
    model = Sequential()
    model.add(Dense(64, input_shape=(X_train.shape[1],), activation='relu'))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))

    # Compile model
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    # Train model
    history = model.fit(X_train[train_index], y_train[train_index], epochs=10, batch_size=32, validation_data=(X_train[val_index], y_train[val_index]))

    # Evaluate model on validation set
    loss, accuracy = model.evaluate(X_train[val_index], y_train[val_index])
    print("Validation Loss:", loss)
    print("Validation Accuracy:", accuracy)
    
    # Store evaluation results
    losses.append(loss)
    accuracies.append(accuracy)

# Calculate average loss and accuracy across all folds
avg_loss = np.mean(losses)
avg_accuracy = np.mean(accuracies)
print("Average Validation Loss:", avg_loss)
print("Average Validation Accuracy:", avg_accuracy)
```

GRID SEARCH

```python
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import GridSearchCV
from scikeras.wrappers import KerasClassifier
import numpy as np

# Define a function to create your Keras model
def create_model(num_layers=2, num_units=8, optimizer='adam', activation='relu'):
    model = Sequential()
    model.add(Dense(num_units, input_shape=(X_train.shape[1],), activation=activation))
    for _ in range(num_layers - 1):  # Adding additional layers
        model.add(Dense(num_units, activation=activation))
    model.add(Dense(1, activation='sigmoid'))  # Output layer
    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Create a KerasClassifier based on your model
keras_model = KerasClassifier(build_fn=create_model, verbose=0)

# Define the hyperparameters grid
param_grid = {
    'num_layers': [1, 2, 3],  # Different numbers of layers to try
    'num_units': [8, 16, 32]  # Different numbers of units per layer to try
}

# Perform Grid Search
grid_search = GridSearchCV(estimator=keras_model, param_grid=param_grid, cv=3, scoring='accuracy')
grid_result = grid_search.fit(X_train, y_train)

# Print best parameters and score
print("Best Parameters:", grid_result.best_params_)
print("Best Accuracy:", grid_result.best_score_)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.model_selection import StratifiedKFold
from scikeras.wrappers import KerasClassifier, KerasRegressor
import numpy as np

# Define a function to create your Keras model
def create_model(optimizer='adam', activation='relu'):
    model = Sequential()
    model.add(Dense(64, input_shape=(X_train.shape[1],), activation=activation))
    model.add(Dense(32, activation=activation))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Create a KerasClassifier based on your model
keras_model = KerasClassifier(build_fn=create_model, epochs=10, batch_size=32, verbose=0)

# Define the hyperparameters grid
param_grid = {
    'optimizer': ['adam', 'sgd'],
    'activation': ['relu', 'sigmoid']
}

# Define the cross-validation strategy
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Perform Grid Search
grid_search = GridSearchCV(estimator=keras_model, param_grid=param_grid, cv=cv, scoring='accuracy')
grid_result = grid_search.fit(X_train, y_train)

# Print best parameters and score
print("Best Parameters:", grid_result.best_params_)
print("Best Accuracy:", grid_result.best_score_)
```
